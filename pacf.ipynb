```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Définition des paramètres de ton modèle MA(2)
# Tu peux modifier ces valeurs (en restant dans le triangle d'inversibilité)
theta1 = 0.6
theta2 = 0.2

# 2. Calcul des autocorrélations théoriques
variance_factor = 1 + theta1**2 + theta2**2
rho_0 = 1.0
rho_1 = (theta1 + theta1 * theta2) / variance_factor
rho_2 = theta2 / variance_factor

# 3. Création du tableau des ACF (jusqu'au lag 10 pour montrer le vide ensuite)
lags = np.arange(11)
acf = np.zeros(11)
acf[0] = rho_0
acf[1] = rho_1
acf[2] = rho_2
# Les lags 3 à 10 restent à 0 (C'est le fameux "cut-off")

# 4. Création de la figure (style publication scientifique)
plt.figure(figsize=(8, 5))

# Création du graphique en bâtons (stem plot)
markerline, stemlines, baseline = plt.stem(lags, acf, basefmt="k-")

# Personnalisation des couleurs et des traits pour un rendu élégant
plt.setp(markerline, marker='o', markersize=8, color='#1f77b4', markeredgecolor='black')
plt.setp(stemlines, color='#1f77b4', linewidth=2)
plt.setp(baseline, color='black', linewidth=1.5)

# Ajout d'une zone grisée (représente souvent l'intervalle de confiance autour de 0)
plt.axhspan(-0.15, 0.15, alpha=0.15, color='gray')

# Titres et labels (en anglais pour ta soutenance)
plt.title(f'Autocorrelation', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Lag ($k$)', fontsize=12)
plt.ylabel('Autocorrelation ($\\rho_k$)', fontsize=12)

# Réglage des axes
plt.xticks(lags)
plt.ylim(-1, 1)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# 5. Sauvegarde de l'image en haute résolution (pour le Beamer LaTeX)
plt.tight_layout()
plt.savefig('ma_plot.png', dpi=300, bbox_inches='tight')

# Affichage du graphique
plt.show()
```
